```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img/duck2.jpeg')
plt.imshow(img)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
low_bound = np.array([170, 120, 70])
up_bound = np.array([180, 255, 255])
```

```python
mask =cv2.inRange(hsv, low_bound, up_bound)
contours, _ = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
for c in contours:
    area = cv2.contourArea(c)
    if area > 2000:
        cv2.drawContours(img, c, -1, (0, 255, 0), 10)
```

```python
cv2.imshow('mask', mask)
cv2.imshow('img', img)
cv2.imwrite('res/red_mask.png', mask)
cv2.imwrite('res/processed_image.png', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
